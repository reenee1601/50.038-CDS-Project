# Understanding the HAM10000 Data

In this step, the team will be viewing and visualising the data to get a better understanding of it. This would help the team to clean and transform the data with more ease.

```python
# imports for understanding data and data visualisation
import pandas as pd
import numpy as np

# data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data/HAM 10000/HAM10000_metadata.csv")

# print data types of variables
print(df.info())

# Display summary statistics
print(df.describe())

# view data frames 
df
```

## Data Cleaning

In this step, the team would perform the following data cleaning techniques:
1. Removing the dx_type column
2. Remove rows with 'unknown' sex and localization, as this data would not be helpful
3. Encoding categorical data for ease of running machine learning algorithms later
4. Removing rows with missing data

A new CSV with cleaned data can then be saved.

```python
# Remove rows with missing data
df_cleaned = df.dropna()

# Remove dx_type column
df_cleaned = df_cleaned.drop(columns=['dx_type'])

# Remove rows with 'unknown' sex and localization
df_cleaned = df_cleaned[df_cleaned['sex'] != 'unknown']
df_cleaned = df_cleaned[df_cleaned['localization'] != 'unknown']

# encode categorical data - sex, localisation and diagnosis
label_encoder_sex = LabelEncoder()
label_encoder_local = LabelEncoder()
label_encoder_dx = LabelEncoder()

df_cleaned['sex'] = label_encoder_sex.fit_transform(df_cleaned['sex'])
df_cleaned['localization'] = label_encoder_local.fit_transform(df_cleaned['localization'])
df_cleaned['dx'] = label_encoder_dx.fit_transform(df_cleaned['dx'])

# Rename columns
df_cleaned.rename(columns={'dx': 'diagnosis'}, inplace=True)

df_cleaned
```

```python
# check how data has been encoded

# For 'sex' column
sex_classes = label_encoder_sex.classes_
print("Encoded values and their corresponding meanings for 'sex' column:")
for i, value in enumerate(sex_classes):
    print(f"Encoded value {i}: {value}")

# For 'anatom_site_general_challenge' column
anatom_site_classes = label_encoder_local.classes_
print("\nEncoded values and their corresponding meanings for 'localization' column:")
for i, value in enumerate(anatom_site_classes):
    print(f"Encoded value {i}: {value}")

# For 'diagnosis' column
diagnosis_classes = label_encoder_dx.classes_
print("\nEncoded values and their corresponding meanings for 'diagnosis' column:")
for i, value in enumerate(diagnosis_classes):
    print(f"Encoded value {i}: {value}")
```

```python
#save df to csv
df_cleaned.to_csv('data/HAM 10000/cleaned_HAM10000_metadata.csv', index=False)
```

## General Findings from Understanding the Data

### Key similarities / differences with ISIC_2020 dataset
- both provide age (age_approx), sex and localization (anatom_site_general challenge) of patient
- HAM10000 categorizes the diagnoses into 7 classes, without indication of benign / malignant.
- by comparison HAM10000 has a smaller size

### Under 'dx', there are 7 classes of skin cancer lesions. They consist of:
- Actinic keratoses and intraepithelial carcinoma / Bowen's disease (akiec)
- basal cell carcinoma (bcc)
- benign keratosis-like lesions (bkl)
- dermatofibroma (df)
- melanoma (mel)
- melanocytic nevi (nv)
- vascular lesions (vas)

## Multivariate Analysis - Exploration of relationship between two or more variables


```python
# Distribution of Age by Diagnosis
plt.figure(figsize=(12, 6))
sns.histplot(data=df_cleaned, x='age', hue='diagnosis', multiple='stack', bins=20, kde=True, palette='viridis')
plt.title('Distribution of Age by Diagnosis')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

<div class="alert alert-block alert-info">
The most common age range in which skin lesions occurs is between ~35 to ~55.
The bellcurve is left skewed, suggesting that skin lesions happens more frequently among older people.
Most common skin lesion is Melanocytic Nevi (NV).
</div>

```python
# Box Plot for Age by Localisation
plt.figure(figsize=(12, 6))
sns.boxplot(x='localization', y='age', data=df_cleaned, palette='viridis')
plt.title('Age Distribution by Localisation')
plt.xlabel('Localisation')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create age groups by grouping ages into every 10 years
df_cleaned['age_group'] = pd.cut(df_cleaned['age'], bins=range(0, 100, 10), right=False)

# Gridplot of Diagnosis by Age Group, 1 Age per plot
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Get unique values in the 'diagnosis' column
unique_age_group = df_cleaned['age_group'].unique()

# Loop through each unique diagnosis value and plot a countplot
for i, age_group_value in enumerate(unique_age_group):
    sns.countplot(x='diagnosis', data=df_cleaned[df_cleaned['age_group'] == age_group_value], ax=axes[i])
    axes[i].set_title(f'Age Group: {age_group_value}')
    axes[i].set_xlabel('Diagnosis')
    axes[i].set_ylabel('Count')

# Hide the unused subplots
for j in range(len(unique_age_group), len(axes)):
    axes[j].axis('off')

plt.suptitle('Gridplot of Diagnosis by Age Group')
plt.tight_layout()
plt.show()

# Distribution of Diagnosis by Age Group
plt.figure(figsize=(12, 6))
sns.countplot(x='age_group', hue='diagnosis', data=df_cleaned, palette='viridis')
plt.title('Diagnosis Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Diagnosis')
plt.show()
```

<div class="alert alert-block alert-info">
Diagnosis 5, Melanocytic Nevi (NV) is prevalent across all the ages. However, Diagnosis 2 and 4, Benign Keratosis-like lesions and Melanoma , seems mostly prevalent in older ages.
</div>

```python
# Distribution of Diagnosis by Gender
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
sns.countplot(x='sex', hue='diagnosis', data=df_cleaned, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Diagnosis by Gender')
plt.legend(title='Diagnosis', loc='upper right')
plt.show()
```

<div class="alert alert-block alert-info">
The diagnosis distribution seems more or less similar for both genders.
</div>

```python
# image visualisation imports

# import system libs
import os
import matplotlib.image as mpimg
import time
import shutil
import itertools

# import data handling tools
import cv2
from tensorflow import keras
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# from sklearn.cluster import KMeans
# from sklearn.decomposition import PCA
# from collections import Counter

# from sklearn.preprocessing import StandardScaler
# from sklearn.manifold import TSNE
```

```python
data_dir = 'data/HAM 10000/hmnist_28_28_RGB.csv'
data = pd.read_csv(data_dir)
data.head()

Label = data["label"]
Data = data.drop(columns=["label"])

oversample = RandomOverSampler()
Data, Label  = oversample.fit_resample(Data, Label)
Data = np.array(Data).reshape(-1, 28, 28, 3)
print('Shape of Data :', Data.shape)

Label = np.array(Label)
Label

classes = {4: ('nv', ' melanocytic nevi'),
           6: ('mel', 'melanoma'),
           2 :('bkl', 'benign keratosis-like lesions'),
           1:('bcc' , ' basal cell carcinoma'),
           5: ('vasc', ' pyogenic granulomas and hemorrhage'),
           0: ('akiec', 'Actinic keratoses and intraepithelial carcinomae'),
           3: ('df', 'dermatofibroma')}

X_train , X_test , y_train , y_test = train_test_split(Data , Label , test_size = 0.25 , random_state = 49)
# print(f'X_train shape: {X_train.shape}\nX_test shape: {X_test.shape}')
# print(f'y_train shape: {y_train.shape}\ny_test shape: {y_test.shape}')

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Create an ImageDataGenerator object
datagen = ImageDataGenerator(
    rotation_range=20,      # Random rotation between 0 and 20 degrees
    width_shift_range=0.2,  # Randomly shift images horizontally (fraction of total width)
    height_shift_range=0.2, # Randomly shift images vertically (fraction of total height)
    horizontal_flip=True,   # Randomly flip images horizontally
    vertical_flip=True,     # Randomly flip images vertically
    shear_range=0.2,        # Shear Intensity
    zoom_range=0.2,         # Range for random zoom
    fill_mode='nearest'     # Fill mode for filling in newly created pixels
)

# Select a random image from X_train
random_index = np.random.randint(0, len(X_train)-1)
random_image = X_train[random_index]

# Reshape the image to (1, height, width, channels) for the datagen
random_image = random_image.reshape((1,) + random_image.shape)

# Generate augmented images and display
plt.figure(figsize=(5,5))
for i, augmented_image in enumerate(datagen.flow(random_image, batch_size=1)):
    if i == 9:
        break
    plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_image[0].astype('uint8'))
    plt.axis('off')
plt.show()
```

```python
# Create a new ImageDataGenerator without augmentation for display
display_datagen = ImageDataGenerator()

# Select a random sample of images from X_train
sample_indices = np.random.choice(len(X_train), size=9, replace=False)
sample_images = X_train[sample_indices]
sample_labels = y_train[sample_indices]

# Reshape the images to (batch_size, height, width, channels) for the datagen
sample_images = sample_images.reshape((-1,) + sample_images.shape[1:])

# Generate augmented images and display with class names
plt.figure(figsize=(6,6))
for i, (augmented_image, label) in enumerate(zip(display_datagen.flow(sample_images, batch_size=1), sample_labels)):
    plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_image[0].astype('uint8'))
    plt.title(classes[np.argmax(label)][1], fontsize=8)  # Get class name from classes dictionary
    plt.axis('off')
    if i == 8:
        break
plt.tight_layout()
plt.show()
```
