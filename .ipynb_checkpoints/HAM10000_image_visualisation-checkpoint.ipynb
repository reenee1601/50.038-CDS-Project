```python
# import system libs
import os
import time
import shutil
import itertools

# import data handling tools
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import random

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from collections import Counter

from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
data_dir = 'data/HAM 10000/hmnist_28_28_RGB.csv'
data = pd.read_csv(data_dir)
data.head()

Label = data["label"]
Data = data.drop(columns=["label"])

oversample = RandomOverSampler()
Data, Label  = oversample.fit_resample(Data, Label)
Data = np.array(Data).reshape(-1, 28, 28, 3)
print('Shape of Data :', Data.shape)

Label = np.array(Label)
Label

classes = {4: ('nv', ' melanocytic nevi'),
           6: ('mel', 'melanoma'),
           2 :('bkl', 'benign keratosis-like lesions'),
           1:('bcc' , ' basal cell carcinoma'),
           5: ('vasc', ' pyogenic granulomas and hemorrhage'),
           0: ('akiec', 'Actinic keratoses and intraepithelial carcinomae'),
           3: ('df', 'dermatofibroma')}

X_train , X_test , y_train , y_test = train_test_split(Data , Label , test_size = 0.25 , random_state = 49)
# print(f'X_train shape: {X_train.shape}\nX_test shape: {X_test.shape}')
# print(f'y_train shape: {y_train.shape}\ny_test shape: {y_test.shape}')

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Create an ImageDataGenerator object
datagen = ImageDataGenerator(
    rotation_range=20,      # Random rotation between 0 and 20 degrees
    width_shift_range=0.2,  # Randomly shift images horizontally (fraction of total width)
    height_shift_range=0.2, # Randomly shift images vertically (fraction of total height)
    horizontal_flip=True,   # Randomly flip images horizontally
    vertical_flip=True,     # Randomly flip images vertically
    shear_range=0.2,        # Shear Intensity
    zoom_range=0.2,         # Range for random zoom
    fill_mode='nearest'     # Fill mode for filling in newly created pixels
)

# Select a random image from X_train
random_index = random.randint(0, len(X_train)-1)
random_image = X_train[random_index]

# Reshape the image to (1, height, width, channels) for the datagen
random_image = random_image.reshape((1,) + random_image.shape)

# Generate augmented images and display
plt.figure(figsize=(5,5))
for i, augmented_image in enumerate(datagen.flow(random_image, batch_size=1)):
    if i == 9:
        break
    plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_image[0].astype('uint8'))
    plt.axis('off')
plt.show()
```

```python
# Create a new ImageDataGenerator without augmentation for display
display_datagen = ImageDataGenerator()

# Select a random sample of images from X_train
sample_indices = np.random.choice(len(X_train), size=9, replace=False)
sample_images = X_train[sample_indices]
sample_labels = y_train[sample_indices]

# Reshape the images to (batch_size, height, width, channels) for the datagen
sample_images = sample_images.reshape((-1,) + sample_images.shape[1:])

# Generate augmented images and display with class names
plt.figure(figsize=(6,6))
for i, (augmented_image, label) in enumerate(zip(display_datagen.flow(sample_images, batch_size=1), sample_labels)):
    plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_image[0].astype('uint8'))
    plt.title(classes[np.argmax(label)][1], fontsize=8)  # Get class name from classes dictionary
    plt.axis('off')
    if i == 8:
        break
plt.tight_layout()
plt.show()
```

```python
# Create a dictionary to store images based on their class
class_images = {class_label: [] for class_label in classes.keys()}

# Group images based on their class
for image, label in zip(X_train, y_train):
    class_images[np.argmax(label)].append(image)

# Plot images for each class in a grid
plt.figure(figsize=(10, 10))

num_images_per_class = 5  # Number of images to display per class
for i, (class_label, images) in enumerate(class_images.items(), 1):
    # Display images for each class
    for j in range(num_images_per_class):
        plt.subplot(len(classes), num_images_per_class, (i - 1) * num_images_per_class + j + 1)
        plt.imshow(images[j])
        plt.title(classes[class_label][0])  # Display short form class name as title
        plt.axis('off')

plt.tight_layout()
plt.show()

# Plot images for each class in a grid
plt.figure(figsize=(10, 10))

num_images_per_class = 5  # Number of images to display per class
for i, (class_label, images) in enumerate(class_images.items(), 1):
    # Display images for each class
    for j in range(num_images_per_class):
        plt.subplot(num_images_per_class, len(classes), j * len(classes) + i)
        plt.imshow(images[j])
        plt.title(classes[class_label][0])  # Display short form class name as title
        plt.axis('off')

plt.tight_layout()
plt.show()

```

```python
# Load your dataset containing RGB pixel values
# Replace 'data.csv' with the actual path to your dataset
data = pd.read_csv('./HAM 10000/hmnist_28_28_RGB.csv')

# Extract class labels from the dataset
class_labels = data['label']  # Assuming the column containing class labels is named 'label'

# Drop the class labels from the dataset
data = data.drop(columns=['label'])

# Apply PCA for dimensionality reduction (optional)
pca = PCA(n_components=2)
data_reduced = pca.fit_transform(data)

# Apply K-means clustering
num_clusters = 7  # Number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(data_reduced)

# Assuming 'classes' is your dictionary of skin lesion classes
classes = {
    4: ('nv', 'melanocytic nevi'),
    6: ('mel', 'melanoma'),
    2: ('bkl', 'benign keratosis-like lesions'),
    1: ('bcc', 'basal cell carcinoma'),
    5: ('vasc', 'pyogenic granulomas and hemorrhage'),
    0: ('akiec', 'Actinic keratoses and intraepithelial carcinomae'),
    3: ('df', 'dermatofibroma')
}

# Map cluster labels to class labels
cluster_labels = []
for cluster_id in range(num_clusters):
    # Get the indices of samples belonging to the current cluster
    cluster_indices = [i for i, cluster_label in enumerate(clusters) if cluster_label == cluster_id]
    # Get the actual class labels of samples in the current cluster
    class_labels_in_cluster = [class_labels[idx] for idx in cluster_indices]
    # Determine the most common class label within the cluster
    most_common_label = Counter(class_labels_in_cluster).most_common(1)[0][0]
    # Map the most common class label to its corresponding class name
    cluster_labels.append(classes[most_common_label][1])

# Visualize clustering results (if applicable)
plt.figure(figsize=(8, 6))
for cluster_id in range(num_clusters):
    cluster_points = data_reduced[clusters == cluster_id]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_id}: {cluster_labels[cluster_id]}')

plt.title('Clustering of Skin Lesions based on RGB Pixel Values')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# Load your dataset containing RGB pixel values
# Replace 'data.csv' with the actual path to your dataset
data = pd.read_csv('./HAM 10000/hmnist_28_28_RGB.csv')

# Extract class labels from the dataset
class_labels = data['label']  # Assuming the column containing class labels is named 'label'

# Drop the class labels from the dataset
data = data.drop(columns=['label'])

# Apply standard scaling to the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
data_reduced_pca = pca.fit_transform(data_scaled)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
data_reduced_tsne = tsne.fit_transform(data_scaled)

# Apply K-means clustering with PCA-reduced data
kmeans_pca = KMeans(n_clusters=5, random_state=42)
clusters_pca = kmeans_pca.fit_predict(data_reduced_pca)

# Apply K-means clustering with t-SNE-reduced data
kmeans_tsne = KMeans(n_clusters=5, random_state=42)
clusters_tsne = kmeans_tsne.fit_predict(data_reduced_tsne)

# Visualize clustering results with PCA-reduced data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for cluster_id in range(5):
    cluster_points = data_reduced_pca[clusters_pca == cluster_id]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_id}')
plt.title('Clustering with PCA-reduced Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

# Visualize clustering results with t-SNE-reduced data
plt.subplot(1, 2, 2)
for cluster_id in range(5):
    cluster_points = data_reduced_tsne[clusters_tsne == cluster_id]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_id}')
plt.title('Clustering with t-SNE-reduced Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()

plt.tight_layout()
plt.show()
```
